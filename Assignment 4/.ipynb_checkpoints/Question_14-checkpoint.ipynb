```python
import numpy as np

np.random.seed(3500)
```

```python
A = []
B = []

success = 0

nsim = 5000
```

```python
for i in range(nsim):
    A.append(np.random.random())
    B.append(np.random.random())

    if abs(A[i] - B[i]) < 0.2:
        success += 1

print("Probability that distance is less than 0.2 = ", success / nsim)
```
