```python
# In this simulation the library numpy is used and it is named it "np"
import numpy as np

# The seed below initializes the random generator to always obtain the same result.
# DO NOT CHANGE THIS VALUE AND RERUN THIS CELL EVERYTIME YOU RUN THE PROGRAM.
np.random.seed(3500)

redTime = 60  # Red light time in seconds
greenTime = 60  # Green light time in seconds
p_restrict = 0.4  # Restrictions will be set in place 40% of the time during construction.
nVehi = 0  # Number of vehicles arriving at the intersection
nRep = 0  # Counter for number of repetitions
nQueue = 0  # Counter for number of times vehicles queue

isRed = True  # Initialize the light as red
isGreen = False

# Monte Carlo simulation
nsim = 80000  # Number of simulation
while nRep < nsim:

    # Restrictions are in place this cycle
    restrict = (np.random.rand() < p_restrict)
    if restrict:
        arrival_rate = 15 / 60  # Rate of vehicles per second with restriction
        queueMoveTime = 1.8  # Time it takes a vehicle to clear the intersection with restriction

    # Restrictions are not in place this cycle
    else:
        arrival_rate = 17 / 60  # Rate without restriction
        queueMoveTime = 1.5  # Time it takes a vehicle to clear the intersection without restriction

    # Red light cycle
    if isRed:
        nRep += 1  # Count replicates
        nVehi += np.random.poisson(arrival_rate * redTime)
        isRed = False
        isGreen = True

    # Green light cycle
    else:
        time = queueMoveTime
        while isGreen:
            if nVehi > 0:
                nVehi -= 1
                nVehi += np.random.poisson(arrival_rate * queueMoveTime)
            else:
                break
            time += queueMoveTime
            if time >= greenTime:
                break

        if nVehi > 0:
            nQueue += 1

        isRed = True
        isGreen = False

# Calculate the probability of vehicles queuing
pQueue_construction = nQueue / nsim
print("The probability of queuing at the intersection is", pQueue_construction)
```
