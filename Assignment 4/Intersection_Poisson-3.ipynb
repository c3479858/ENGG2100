### Intersection Poisson


A single lane of traffic has to cross a busy intersection. Traffic lights are to be installed at the intersection. There is some concern about the ability of the traffic lights to cope with the expected traffic volume.

To better understand the dynamics of the intersection it is proposed that the problem should be analyzed using a Monte Carlo program to evaluate the performance of the intersection under different traffic loads. 




The following considerations are made:

        - The arrival of vehicles at the intersection follows a Poisson process with the parameter v being the rate of vehicles per second.
        - The red and green light times are 60 seconds. Yellow light and all-red times can be ignored.
        - If vehicles have queued at the intersection during a red cycle, it takes each vehicle at head of the queue 1.5 seconds to cross or turn at the intersection once the lights change to green.



Below is a Monte Carlo problem that simulates the engineered system and calculates the probability of queuing at the intersection given a set configuration of Red and Green light times as well as rate of vehicles arrival.




```python
# In this simulation the library numpy is used and it is named it "np"

import numpy as np

# The seed below initializes the random generator to always obtain the same result. 
# DO NOT CHANGE THIS VALUE AND RERUN THIS CELL EVERYTIME YOU RUN THE PROGRAM. 
np.random.seed(3500)    
```

```python
# Variables are defined in this cell 

v = 15 / 60  # Rate of vehicles per second
queueMoveTime = 1.8  # Time it takes a vehicle to clear the intersection

redTime = 60  # Red light time in seconds
greenTime = 60  # Green light time in seconds

nVehi = 0  # Number of vehicles arriving at the intersection
nRep = 0  # Counter for number of repetitions
nQueue = 0  # Counter for number of times vehicles queue

isRed = True  # Initialize the light as red
isGreen = False

# Monte Carlo simulation
nsim = 100000   # Number of simulation                   
```

## Main loop of the simulation 



```python

while nRep < nsim:
    # Red light cycle
    if isRed:
        nRep = nRep + 1  # Count replicates
        nVehi = nVehi + np.random.poisson(v * redTime)

        isRed = False
        isGreen = True

    # Green light cycle
    else:
        time = queueMoveTime
        while isGreen:
            if nVehi > 0:
                nVehi = nVehi - 1
                nVehi = nVehi + np.random.poisson(v * queueMoveTime)
            else:
                break
            time = time + queueMoveTime
            if time >= greenTime:
                break

        if nVehi > 0:
            nQueue = nQueue + 1

        # Reset light to Red
        isRed = True
        isGreen = False

# Calculate the probability of queuing
pQueue = nQueue / nsim

print("The probability of queuing at the intersection is ", pQueue)
```
